Checking Skewness for all columns

```python
plt.figure(figsize = (20,15),facecolor = 'red')
plotnumber = 1
for column in df:
    if plotnumber <= 15:
        ax = plt.subplot(5,3,plotnumber)
        sns.distplot(df[column])
        plt.xlabel(column,fontsize = 20)
    plotnumber +=1
plt.show()
```

We can note all columns except target column has skewness due to outliers

z-score or vif can be used to remove this skewness by removing outliers

We need to turn these columns from skewedly normally distributed to normally distributed columns or z-distributed
